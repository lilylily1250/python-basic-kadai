```python
#Humanのclassを定義する
class Human:

  #コンストラクタを定義する
  def __init__(self, name, age):

    #属性nameを定義する
    self.name = name

    #属性ageを定義する
    self.age = age

  #メソッドを定義する
  def printinfo(self):
    print(self.name)
    print(self.age)

#インスタンス化し、humanという変数に代入する
#つまりhumanというオブジェクトを作成する
human = Human("侍太郎", 36)

#属性にアクセスし、値を出力する
human.printinfo()

```
